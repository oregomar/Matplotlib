```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make data
x = np.linspace(-4, 4, 6)
y = np.linspace(-4, 4, 6)

X, Y = np.meshgrid( x, y )

U = 0.8*Y + 0.2*X + 0.6*Y + 0.5*X
V = 0.2*Y - 0.8*X + 0.5*Y - 0.6*X

U=-U
V=-V

# plot
fig, ax = plt.subplots()

ax.quiver(X, Y, U, V, color="C0", angles='xy', scale_units='xy', scale=5, width=.005)

#ax.set(xlim=(-5, 5), ylim=(-5, 5))

plt.show()

# plot
fig, ax = plt.subplots()

ax.streamplot(X, Y, U, V)

plt.show()
```
