```python
import matplotlib.pyplot as plt
import numpy as np

#plt.style.use('_mpl-gallery-nogrid')

# make data
x = np.linspace(-10, 10, 11)
y = np.linspace(-10, 10, 11)

X, Y = np.meshgrid( x, y )

U = 0.8*Y + 0.2*X + 0.6*Y + 0.5*X
V = 0.2*Y - 0.8*X + 0.5*Y - 0.6*X

# plot
fig, ax = plt.subplots(1,2,figsize=(6,4))

ax[0].quiver(X, Y, -U, -V, color="C0", angles='xy', scale_units='xy', scale=5, width=.005)
ax[1].streamplot(X, Y, -U, -V)

plt.show()
```
