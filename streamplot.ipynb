
# streamplot(X, Y, U, V)

See `~matplotlib.axes.Axes.streamplot`.


```python
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d.axes3d import Axes3D
import numpy as np

#plt.style.use('_mpl-gallery-nogrid')

# make a stream function:
X, Y = np.meshgrid(np.linspace(-3, 3, 601), np.linspace(-3, 3, 601))

#Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# make U and V out of the streamfunction:
V = np.diff(Z[1:, :], n=1, axis=1)
U = -np.diff(Z[ :, 1:], n=1,  axis=0)

# plot:
fig, ax = plt.subplots(1,2,figsize=(16,8))
ax[0].contourf(X, Y, Z, cmap=cm.coolwarm)
ax[0].streamplot(X[1:, 1:], Y[1:, 1:], U, V, color="k", density=1)

ax=fig.add_subplot(1, 2, 2, projection='3d')
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)
#ax.plot_wireframe(X, Y, Z,  color='k')
ax.view_init(45, 250)
plt.show()
```
