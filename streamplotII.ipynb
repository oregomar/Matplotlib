
# streamplot(X, Y, U, V)

See `~matplotlib.axes.Axes.streamplot`.


```python
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d.axes3d import Axes3D
import numpy as np

#plt.style.use('_mpl-gallery-nogrid')

# make a stream function:
X, Y = np.meshgrid(np.linspace(-3, 3, 252), np.linspace(-3, 3, 252))

#Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

R = np.sqrt(X**2 + Y**2)
#Z = np.sin (R)
#Z = R*np.exp(-R**2)
#Z =  np.sin(R)- np.exp(-R**2)
Z = 180/(R**2+40) -  np.exp(- R**2 )

# make U and V out of the streamfunction:

#divergencia
V = np.diff(Z[:, 1:],  axis=0)
U = np.diff(Z[1:,:],  axis=1)

# plot:
fig, ax = plt.subplots(1,2,figsize=(12,6))
ax[0].contourf(X, Y, Z, cmap=cm.coolwarm)
ax[0].streamplot(X[1:, 1:], Y[1:, 1:], U, V, color="k", density=1)
ax[0].set_title("Divergencia")

Va = np.sqrt(V**2 + U**2)

ax=fig.add_subplot(1, 2, 2, projection='3d')
ax.plot_surface(X[1:, 1:], Y[1:, 1:], Va, cmap=cm.coolwarm)
ax.view_init(65, 250)
plt.show()

#vorticidad
V = -np.diff(Z[1:, :],  axis=1)
U = np.diff(Z[:,1:],  axis=0)

# plot:
fig, ax = plt.subplots(1,2,figsize=(12,6))
ax[0].contourf(X, Y, Z, cmap=cm.coolwarm)
ax[0].streamplot(X[1:, 1:], Y[1:, 1:], U, V, color="k", density=1)
ax[0].set_title("Vorticidad")

Va = np.sqrt(V**2 + U**2)

ax=fig.add_subplot(1, 2, 2, projection='3d')
ax.plot_surface(X[1:, 1:], Y[1:, 1:], Va, cmap=cm.coolwarm)
ax.view_init(55, 240)
plt.show()
V = -np.diff(Z[1:,:], axis=1) + np.diff(Z[:,1:], axis=0)
U =  np.diff(Z[:,1:], axis=0) + np.diff(Z[1:,:], axis=1)

# plot:
fig, ax = plt.subplots(1,2,figsize=(12,6))
ax[0].contourf(X, Y, Z, cmap=cm.coolwarm)
ax[0].streamplot(X[1:, 1:], Y[1:, 1:], U, V, color="k", density=1)
ax[0].set_title("Remolinos")

Va = np.sqrt(V**2 + U**2)

ax=fig.add_subplot(1, 2, 2, projection='3d')
ax.plot_surface(X[1:, 1:], Y[1:, 1:], Va, cmap=cm.coolwarm)
ax.view_init(45, 230)
plt.show()

fig, ax = plt.subplots(1,2,figsize=(12,6))
ax=fig.add_subplot(1, 2, 1, projection='3d')
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)
ax.view_init(45, 230)

ax=fig.add_subplot(1, 2, 2, projection='3d')
ax.plot_surface(X[1:, 1:], Y[1:, 1:], Va, cmap=cm.coolwarm)
ax.view_init(65, 230)
plt.show()

```
